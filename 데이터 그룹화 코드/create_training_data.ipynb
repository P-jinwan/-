# `4.` 특수문자 제거

```python
# 관광지 데이터의 특수문자를 제거합니다.
result = [] # replace 함수로 데이터의 특수문자를 제거하고 결과물을 받을 리스트 변수입니다.

for i in tourist_information['테마'] : # tourist_information['테마']의 샘플 개수만큼 반복합니다.
  i = i.replace(' ', '')               # 문장의 공백을 제거합니다.
  i = i.replace('\n','')               # 이스케이프 문자인 \n(줄바꿈)을 제거합니다.
  result.append(i.split('#')[1:])      # "#(해시 기호)"를 기준으로 각 문장의 인덱스 번호 1번부터 끝까지 나눕니다. (#섬속의섬, #경관/포토 → 섬속의섬, 경관/포토)

# 위에서 분리한 데이터를 tourist_information['테마'] 컬럼의 새로운 데이터로 대입합니다.
tourist_information['테마'] = result
```

## `5-2.` 고유 테마명 추출

## `5-3.` 검증

```python
def ValidateThemeName(list_theme_names) :
  for i in list_theme_names :                            # 테마명이 정의된 리스트 변수를 하나씩 넘겨주며 반복합니다.
    if i in all_themes :                                 # 테마 고유명 전체가 든 리스트 변수에 있다면
      print("테마명이 올바릅니다.")                      # "테마명이 올바릅니다."를 출력합니다.
    else :                                               # 없다면
      print('테마명이 올바르지않습니다. ({})'.format(i)) # '테마명이 올바르지않습니다.'를 출력합니다.

# 반복문으로 전체 테마 리스트를 함수로 보내 테마명이 올바른지 검증합니다.
for i in all_theme_list :
  ValidateThemeName(i)
```

# `6.` 유의어 하나로 묶기

```python
# 바꿔야 하는 값을 정의합니다.
list_elements_replace = [['문화', '문화관광', '문화유적지'], ['4.3', '제주4.3', '너븐숭이', '북촌'], ['신나락만나락', '신화탐방로'], 
                         ['역사유적', '토기', '유적'], ['일몰', '일몰명소', '일출'],
                         ['경관/포토', '생태', '생태문화', '세계자연유산', '자연경관', '유네스코'], ['야영장', '캠핑장', '캠핑', '글램핑'],
                         ['제주동백수목원', '동백나무', '동백'], ['도보', '도보여행'], ['마을관광', '마을산책'], 
                         ['미술관', '미술/박물관', '전시', '전시/행사', '전시관', '전시와행사'],  ['미술', '예술', '미디어아트', '아트'],
                         ['걷기/등산', '내창트레킹', '등산', '트레킹', '지질트레일'], ['걷기', '해안산책'],
                         ['체험관광', '체험여행', '체험', '농촌체험마을', '동물먹이주기'],
                         ['플라잉요가', '하타요가', '메디컬요가', '빈야사요가', '요가', '테랍툴요가'], ['테마파크', '테마공원', '어트랙션'],
                         ['waterleisure', '수상레저'], ['마린스포츠', '물놀이', '수영장'],
                         ['스쿠버다이빙', '스쿠버교육', '다이빙체험'], ['어린이', '아이'], ['사찰', '절'], ['낚시', '낚시체험'], ['사게절', '사계절']]

# 바꿀 값을 정의합니다.
value_replaced = ['문화', '4.3', '신화', '유적', '일출몰', '자연경관', '캠핑장', '동백', '도보', '마을', '미술전시', '예술', '등산', '걷기', '체험', '요가', '테마파크', '수상 스포츠', '수영',
                  '스쿠버다이빙', '어린이', '절', '낚시', '사계절']

# tourist_information['테마'] 컬럼을 theme_copy 변수에 복사하여 시리즈 타입으로 대입합니다.
theme_copy = tourist_information['테마'].copy()

# 위에서 정의한 데이터를 바탕으로 유의어를 하나의 단어로 치환합니다.
for i in range(len(theme_copy)) :                       # theme_copy 변수의 저장된 시리즈 샘플 수만큼 반복합니다. (1,096)
  for j in range(len(theme_copy[i])) :                  # theme_copy 시리즈의 [i] 번째 리스트 인덱스 수만큼 반복합니다. (i: 0, j: 0)
    for k in range(len(value_replaced)) :               # 바꿀 값의 개수 만큼 반복합니다. (i: 0, j: 0, k: 0)
      if theme_copy[i][j] in list_elements_replace[k] : # theme_copy[i][j] 번째 값이 list_elements_replace[k] 번째 값에 있다면 
        theme_copy[i][j] = value_replaced[k]            # theme_copy[i][j] 번째 값을 value_replaced[k] 번째 값으로 바꿉니다.
  theme_copy[i] = list(set(theme_copy[i]))              # 중복값을 없애고 반환합니다.

# tourist_information['테마'] 컬럼의 데이터를 theme_copy 데이터로 치환합니다.
tourist_information['테마'] = theme_copy

# 데이터를 확인합니다.
display(tourist_information.head())
```
